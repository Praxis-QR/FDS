## Data Cleaning

### Convert Year to string

```
movies['Year'].unique()
```

```
len(movies['Year'].unique())
```

```
movies.info()
```

```
movies['Year'] = movies['Year'].astype(str)
```

```
#movies.info()
```

```
movies['Year'].unique()
```

```
movies['Year'] = movies['Year'].str[:4]
movies['Year'].unique()
```

```
movies.isna().sum()
```

```
movies.loc[movies['Year'] == 'nan','Year'] = '2999'
```

```
movies['Year'].unique()
```

```
movies.info()
```

### Convert Reviews to Integer

```
movies.fillna({'Reviews by Users':0.0,'Reviews by Crtiics':0.0}, inplace=True)
```

```
movies = movies.astype({"Reviews by Users":"int","Reviews by Crtiics":"int"})
```

```
movies.info()
```

```
movies.head(2)
```

### Group BY vs Pivot

```
dfGE_01 = movies[['Year', 'Gross Earnings']].groupby(['Year']).sum(numeric_only=True).astype('int64')
dfGE_01
```

```
#movies[['Year', 'Gross Earnings']].groupby(['Year']).sum(numeric_only=True).astype('int64').plot()
dfGE_01.plot()
```

```
#dfGE_02 = movies[['Year', 'Gross Earnings']].pivot_table(index=['Year']).astype('int64')
dfGE_02 = pd.pivot_table(movies, index = 'Year', values = 'Gross Earnings', aggfunc = 'sum').astype('int64')
dfGE_02
```

```
#movies[['Year', 'Gross Earnings']].pivot_table(index=['Year']).astype('int64').plot()
dfGE_02.plot()
```

```
# Pandas.merge() by Column
dfGE=dfGE_01.join(dfGE_02, lsuffix="_GroupBy", rsuffix="_PivotTable", how='outer')
dfGE
```

```
import pandas as pd
import xlsxwriter


# Create a Pandas Excel writer using XlsxWriter as the engine.
writer3 = pd.ExcelWriter('Movies3.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
movies.to_excel(writer3, sheet_name='movies3')
# Close the Pandas Excel writer and output the Excel file.
writer3.close()
```

```
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer32 = pd.ExcelWriter('Movies32.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
movies.to_excel(writer32, sheet_name='movies32')
# Close the Pandas Excel writer and output the Excel file.
writer32.close()
```

```
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer33 = pd.ExcelWriter('Movies33.xlsx', engine='xlsxwriter')


# Turn off the default header and skip one row to allow us to insert a
# user defined header.
movies.to_excel(writer33, sheet_name='Movies33', startrow=1, header=False)

# Get the xlsxwriter workbook and worksheet objects.
workbook  = writer33.book
worksheet = writer33.sheets['Movies33']

# Add a header format.
header_format = workbook.add_format({
    'bold': True,
    'text_wrap': True,
    'valign': 'top',
    'fg_color': '#D7E4BC',
    'border': 1})

# Write the column headers with the defined format.
for col_num, value in enumerate(movies.columns.values):
    worksheet.write(0, col_num + 1, value, header_format)

# Close the Pandas Excel writer and output the Excel file.
writer33.close()
```
