#Imports

```
import pandas as pd
import numpy as np
```

#Read Data

```
# Data provided by https://www.superdatascience.com/
df = pd.read_csv("OfficeSupplies.csv")
```

#General Analysis

```
df.shape
```

```
df.dtypes
```

```
df.head()
```

#Any null values in the dataset?

```
#df.isnull().values.any()
df.isnull().values.any()
```

#Reps who sold the most units

```
#df.groupby(["Rep"]).sum().sort_values("Units", ascending=False)
#df.groupby(["Rep"]).sum().sort_values("Units")
#df.groupby(["Rep"]).sum()
df.groupby(["Rep"]).sum(numeric_only=True)
#df.groupby(["Rep"]).mean()
```

#Add a new column to our dataset

```
df["Value"] = df["Units"] * df["Unit Price"]
df.head()
```

#Pivot Table

```
pd.pivot_table(df[['Region','Item','Units','Value']],index="Region", columns="Item", values = 'Value',aggfunc="sum")
```

```
pd.pivot_table(df[['Region','Item','Units','Value']],index="Region", columns="Item", values = 'Value', margins =True,aggfunc="sum")
```

#Additional analysis

```
#
# Who sold the most?
#
#f.groupby("Rep").sum().sort_values("Total Price", ascending=False).head()
df.groupby("Rep").sum(numeric_only=True).sort_values("Value", ascending=False).head()
#df.groupby("Rep").sum()[['Units','Total Price']].sort_values("Total Price", ascending=False).head()
```

```
#
# Who sold the most?
#
#df.groupby("Rep").sum().sort_values("Total Price", ascending=False).head()
df.groupby("Rep").sum()[['Units','Value']].sort_values("Value", ascending=False).head()
```

```
#
# Who sold most units
#
df.groupby("Rep").sum()[['Units','Value']].sort_values("Units", ascending=False).head()
```

```
#
# Which region sold the most?
#
df.groupby("Region")[['Units','Value']].sum()
```

```
#
# What item sold the most
#
df.groupby("Item")[['Units','Value']].sum()
```

```
#
# who sold the most in each region
#
group = df.groupby(["Region","Rep"]).sum()
total_price = group["Value"].groupby(level=0, group_keys=False)
total_price.nlargest(5)
```

```
df.groupby(["Region","Rep"])[['Units','Value']].sum()
```
