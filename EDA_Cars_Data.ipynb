#Setup, Imports 

```python
# Importing required libraries.
import pandas as pd
import numpy as np
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
#%matplotlib inline 
sns.set(color_codes=True)
```

```python
!wget -O carsData.zip -q 'https://github.com/Praxis-QR/FDS/blob/main/data/cars.zip?raw=true'
```

```python
dfCars = pd.read_csv('carsData.zip')
dfCars.head()
```

```python
# To display the bottom 5 rows
dfCars.tail(5) 
```

#Analysis

```python
# Dropping irrelevant columns
dfCars = dfCars.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
dfCars.head(5)
```

```python
# Renaming the column names
dfCars = dfCars.rename(columns={'Engine HP': 'HP', 'Engine Cylinders': 'Cylinders', 'Transmission Type': 'Transmission', 'Driven_Wheels': 'Drive Mode','highway MPG': 'MPG-H', 'city mpg': 'MPG-C', 'MSRP': 'Price' })
dfCars.head(5)
```

## Removing Duplicates

```python
# Total number of rows and columns
dfCars.shape
```

```python
# Rows containing duplicate data
duplicate_rows_df = dfCars[dfCars.duplicated()]
print('number of duplicate rows: ', duplicate_rows_df.shape)
```

```python
# Dropping the duplicates 
dfCars = dfCars.drop_duplicates()
dfCars.shape
```

```python
# Counting the number of rows after removing duplicates.
dfCars.count()
```
