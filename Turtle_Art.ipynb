```python
from Turtle import *
import random
```

```python
def chooseColour():
    #VALID_COLORS = ('white', 'yellow', 'orange', 'red', 'green', 'blue', 'purple', 'grey', 'black')
    c = random.randint(0,6)
    if c == 0:
        return 'white'
    elif c == 1:
        return 'yellow'
    elif c == 2:
        return 'orange'
    elif c == 3:
        return 'red'
    elif c == 4:
        return 'green'
    elif c == 5:
        return 'blue'
    elif c == 6:
        return 'purple'
```

```python
def cleanSlate(startX, startY):
    initializeTurtle(initial_window_size=(900,900))
    bgcolor('black')
    color('black')
    penup()
    goto(startX, startY)
    pendown()
```

```python
#bgcolor('black')
cleanSlate(450,450)
#turtle.colormode(255)
speed(10)
for x in range(500): 
    #r,b,g=random.randint(0,255),random.randint(0,255),random.randint(0,255)
    #pencolor = (r,g,b)
    color(chooseColour())
    forward(x+50)
    right(91)
#turtle.exitonclick()
```
