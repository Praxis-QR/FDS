```python
import pandas as pd
```

```python
# Load Data

movies_df = pd.read_csv("movies.csv", index_col=False)
movies_df.head(3)
```

```python
movies_df.info()
```

```python
# See Columns
movies_df.columns
```

```python
movies_df.shape
```

```python
movies_df.drop(['Poster_Link','Overview'], axis=1, inplace=True)
movies_df.columns
```

```python
movies_df.shape
```

```python
movies_df.info()
```

```python
movies_df.head(3)
```

```python
# Remove commas and convert Gross to numeric
movies_df['Gross'] = movies_df['Gross'].str.replace(',', '', regex=True)
movies_df['Gross'] = pd.to_numeric(movies_df['Gross'], errors='coerce')
```

```python
# Remove commas and convert Gross to numeric
movies_df['Runtime'] = movies_df['Runtime'].str.replace(' min', '', regex=True)
movies_df['Runtime'] = pd.to_numeric(movies_df['Runtime'], errors='coerce')
```

```python
movies_df['Released_Year'] = pd.to_numeric(movies_df['Released_Year'], errors='coerce').astype('Int64')
```

```python
movies_df.info()
```

```python
movies_df.isnull().sum()
```

```python
#1. Top 5 highest rated movies
#Task: Find the top 5 movies with the highest IMDB rating.
movies_df.nlargest(5, "IMDB_Rating")[["Series_Title", "IMDB_Rating"]]
```

```python
#2. Longest movies
#Task: Find the top 5 longest movies by runtime.
movies_df.nlargest(5, "Runtime")[["Series_Title", "Runtime"]]
```

```python
#3. Highest grossing movies
# Task: List the top 5 movies by box office gross.
movies_df.nlargest(5, "Gross")[["Series_Title", "Gross"]]
```

```python
#4. Average rating
#Task: What is the average IMDB rating of all movies?
print(movies_df["IMDB_Rating"].mean())
```

```python
#5. Most common certificate
#Task: Which certificate (U, UA, A, etc.) is most common?
movies_df["Certificate"].mode()[0]
```

```python
movies_df["Certificate"].value_counts()
```

```python
movies_df["Genre"].value_counts()
```

```python
#7. Most prolific director
# Task: Which director has the most movies in this dataset?
movies_df["Director"].value_counts().head(3)
```

```python
#8. Most frequent lead actor
#Task: Who appears most often as Star1?
movies_df["Star1"].value_counts().head(3)
```

```python
# Combine all actor columns into one series and count frequency
actor_counts = pd.concat([movies_df["Star1"], movies_df["Star2"], movies_df["Star3"], movies_df["Star4"]]).value_counts()

# Most frequent actor
most_frequent_actor = actor_counts.head(3)

print(most_frequent_actor)
```

```python
#9. Top directors by gross
#Task: Find the top 5 directors by average box office gross.
movies_df.groupby("Director")["Gross"].mean().sort_values(ascending=False).head(5)
```

```python
#10. Christopher Nolan’s top movie
# Task: Which Christopher Nolan movie had the highest gross?
movies_df[movies_df["Director"]=="Christopher Nolan"].nlargest(5, "Gross")[["Series_Title", "Gross"]]
```

```python
movies_df[movies_df["Director"]=="Christopher Nolan"][["Series_Title", "Gross"]]
```

```python
#11. Al Pacino’s best film
#Task: Which movie with Al Pacino has the highest IMDB rating?

movies_df[(movies_df[['Star1','Star2','Star3','Star4']] == 'Al Pacino').any(axis=1)] \
  .nlargest(1, "IMDB_Rating")[["Series_Title", "IMDB_Rating"]]
```

```python
#11. Al Pacino’s best film
#Task: Which movie with Al Pacino has the highest IMDB rating?

movies_df[(movies_df[['Star1','Star2','Star3','Star4']] == 'Al Pacino').any(axis=1)] \
  .nlargest(1, "Gross")[["Series_Title", "IMDB_Rating", "Gross"]]
```

```python
#12. Rating by certificate
#Task: What is the average IMDB rating for each certificate?
movies_df.groupby("Certificate")["IMDB_Rating"].mean()
```

```python
#13. Most voted movies
#Task: Which 5 movies got the highest number of votes?
movies_df.nlargest(5, "No_of_Votes")[["Series_Title", "No_of_Votes"]]
```

```python
movies_df.columns
```

```python
#14. Oldest movie
#Task: What is the oldest movie in the dataset?
movies_df.nsmallest(1, "Released_Year")[["Series_Title", "Released_Year"]]
```

```python
movies_df.nlargest(1, "Released_Year")[["Series_Title", "Released_Year"]]
```

```python
#16. Best director (min 5 films)
#Task: Which director has the highest average rating (only if they have 5+ films)?
movies_df.groupby("Director").filter(lambda x: len(x) >= 5) \
  .groupby("Director")["IMDB_Rating"].mean().sort_values(ascending=False).head(3)
```

```python
movies_df.groupby("Director")['Series_Title'].count()
```

```python
movies_df.groupby("Director").filter(lambda x: len(x) >= 5)
```

```python
#17. Average gross (last 10 years)
#Task: Show the average gross per year for the last 10 years in the dataset.
recent_years = movies_df["Released_Year"].max() - 10
recent_years
```

```python
movies_df[movies_df["Released_Year"] >= recent_years].groupby("Released_Year")["Gross"].mean()
```

```python
#18. Most common genres
#Task: What are the top 10 most frequent genres?
movies_df["Genre"].value_counts().head(10)
```

```python
#19. Best movie per decade
#Task: Find the highest rated movie of each decade.
movies_df["Decade"] = (movies_df["Released_Year"] // 10) * 10
movies_df.loc[movies_df.groupby("Decade")["IMDB_Rating"].idxmax()][["Decade", "Series_Title", "IMDB_Rating"]]
```

```python
#20. Correlation between rating & gross
#Task: Is there a correlation between IMDB ratings and box office gross?
movies_df["IMDB_Rating"].corr(movies_df["Gross"])
```
