```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df
```

```python
#Dropping columns in a DataFrame
df.drop(['Unnamed: 0'], axis = 1, inplace=True)
df
```

```python
#Drop rows with blanks
df2a = df.dropna()
df2a
```

```python
#Replace null values with computed values
x = df["Calories"].median()
df2b = df
df2b["Calories"].fillna(x, inplace = True)
df2b
```

```python
df2b.info()
```

```python
#Fixing Date
df3 = df2b
df3['Date'] = pd.to_datetime(df3['Date'])
df3
```
