```
import numpy as np
import pandas as pd
```

```
#Read the Data
theData = pd.read_csv('FAO.csv', encoding='latin')
#theData = pd.read_csv('FAO.csv', encoding='utf-8')
#theData = pd.read_csv('FAO.csv')
```

#Reduce the Data

```
#Get Column Names
theData.columns
```

```
IndiaData = theData[theData['Area']== 'India']
IndiaData
```

```
Food = theData[theData['Element']== 'Food']
Food
```

```
IndiaFood = theData[(theData['Area'] == 'India') & (theData['Element']== 'Food')]
IndiaFood
```

```
UseLessColumns = ['Area Abbreviation', 'Area Code','latitude','longitude']
YearlyData = theData.drop(columns = UseLessColumns)
YearlyData
```

```
IndiaFood = YearlyData[(YearlyData['Area'] == 'India') & (YearlyData['Element']== 'Food')]
IndiaFood
```

```
DropColumns = ['Area','Item Code','Element Code','Element','Unit']
IndiaFood2 = IndiaFood.drop(columns = DropColumns)
IndiaFood2
```

```
IndiaFood2.index = IndiaFood2.Item
IndiaFood2
```

```
IndiaFood3 = IndiaFood2.drop(columns ='Item')
IndiaFood3
```

#Transpose (Rotate) the Data

```
IndiaFoodTimeSeries = IndiaFood3.T
IndiaFoodTimeSeries
```

#Plot the Data

```
IndiaFoodTimeSeries.plot()
```

```

IndiaFoodTimeSeries[['Oilcrops','Animal fats']].plot()
```

```
IndiaFoodTimeSeries[['Cephalopods','Maize and products']].plot()
```

```
ProductsOfInterest = ['Wheat and products','Maize and products']
IndiaFoodTimeSeries[ProductsOfInterest].plot(figsize=(12,8))
```

#Statistical Analysis <br>
[Chris Albon Tutorial](https://chrisalbon.com/python/data_wrangling/pandas_dataframe_descriptive_stats/)

```
#Reduce the number of columns
ColumnsOfInterest = ['Wheat and products','Rice (Milled Equivalent)','Potatoes and products','Coconut Oil','Alcoholic Beverages']
IFTS2 = IndiaFoodTimeSeries[ColumnsOfInterest]
IFTS2
```

```
#Smaller, convenient column names
IFTS2.rename(columns={"Wheat and products": "Wheat"}, inplace=True)
IFTS2.rename(columns={"Rice (Milled Equivalent)": "Rice"}, inplace=True)
IFTS2.rename(columns={"Potatoes and products": "Potato"}, inplace=True)
IFTS2.rename(columns={"Coconut Oil": "Coconut"}, inplace=True)
IFTS2.rename(columns={"Alcoholic Beverages": "Alcohol"}, inplace=True)
IFTS2
```
