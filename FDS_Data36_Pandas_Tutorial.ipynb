#Tutorial 2

```
dataURL = 'https://raw.githubusercontent.com/Praxis-QR/FDS/main/data/zoo.csv'
dfZoo = pd.read_csv(dataURL)
dfZoo.head()
```

##Data Aggregation

```
dfZoo.count()
```

```
dfZoo[['animal']].count()
```

```
dfZoo.animal.count()
```

```
dfZoo.water_need.sum()
```

```
dfZoo.sum()
```

```
dfZoo.water_need.min()
```

```
dfZoo.water_need.mean()
```

#Grouping

```
dfZoo.groupby('animal').mean()
```

```
dfZoo.groupby('animal').mean()[['water_need']]
```

```
dfZoo.groupby('animal').mean().water_need
```

```
df_ArticlesRead.groupby('source').count()
```

```
df_ArticlesRead.groupby('source').count()[['user_id']]

```

```
df_ArticlesRead[df_ArticlesRead.country == 'country_2'].groupby(['source', 'topic']).count()
```
