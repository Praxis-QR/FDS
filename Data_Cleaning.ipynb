```python
#Importing the dataset by reading the csv file
#df = pd.read_csv('AB_NYC_2019.csv')
df = pd.read_csv('AB_NYC.csv')
df.head()
```

```python
df.info()
```

```python
#Finding duplicated values in a DataFrame
df.duplicated()
```

```python
#Finding duplicated values in a DataFrame
df.duplicated().any()
```

```python
#df.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
df.drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=False)
```

```python
#Finding duplicated values in a DataFrame
df.duplicated().any()
```

```python
#Finding missing values in a DataFrame
df.isnull()
```

```python
#Finding missing values in a DataFrame
df.isna()
```

```python
#Finding missing values in a DataFrame
df.isna().any()
```

```python
#Finding missing values in a DataFrame
df.isna().sum()
```

```python
#Replacing N/A values in a DataFrame
#df.fillna({'name':'Not Stated','host_name':'Not Stated','last_review':0}, inplace=True)
df.fillna({'name':'Not Stated','host_name':'Not Stated','last_review':'1-1-1900'}, inplace=True)
```

```python
#19-10-2018
```

```python
df.isna().sum()
```

```python
#Dropping columns in a DataFrame
df.drop(['reviews_per_month'], axis = 1, inplace=True)
```

```python
df.columns
```

```python
#Changing the index of a DataFrame
df['id'].is_unique
```

```python
#Let’s set the 'id' column as index using set_index:
df = df.set_index('id')
df.head()
```

```python
df.iloc[4]
```

```python
#Renaming column of a DataFrame
new_col = {'name':'listing_name', 'number_of_reviews':'reviews'}

df.rename(columns=new_col, inplace=True)
df.head()
```

```python
df['last_review'].dtype.type
```

```python
#Converting data type
#df['last_review'] = pd.to_datetime(df['last_review'], format='%Y-%m-%d')
#df['last_review'].dtype.type
```

```python
#Converting data type
df['last_review'] = pd.to_datetime(df['last_review'], format='%d-%m-%Y')
df['last_review'].dtype.type
```

```python
#Converting data type to reduce memory
df['host_id'] = df['host_id'].astype('int32')
df.info()
```
