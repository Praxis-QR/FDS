```python
import folium
import pandas as pd
import osmnx as ox
```

```python
#place = 'Manhattan, New York, USA'
place = 'Kolkata, West Bengal, India'
```

#Points of Interest

```python
#tags = {'amenity': 'cafe'}                                      # https://wiki.openstreetmap.org/wiki/Map_features
#tags = {'amenity': 'restaurant', 'amenity': 'cafe'}               this does not work
tags = {'amenity': 'restaurant'} 
poi_s = ox.geometries_from_place(place, tags=tags)
poi_s.sample(5)
#poi_s[poi_s['amenity'] == 'cafe']
```

```python
#poi_s.type
poi_s_points = poi_s[poi_s.geom_type == 'Point'][:100]
```

```python
#m = folium.Map([40.754932, -73.984016], zoom_start=10)
m = folium.Map([22.554027, 88.351393], zoom_start=16)                           # Kolkata, Park Hotel
locs = zip(poi_s_points.geometry.y, poi_s_points.geometry.x)
for location in locs:
    folium.CircleMarker(location=location).add_to(m)
    m.save('map01.html')
m
```

#Streets

```python
#place = 'Manhattan, New York, USA'
place = 'Kolkata, West Bengal, India'
#graph = ox.graph_from_place(place, network_type='drive')
location_point = (22.5176397,88.3658511)    # Gariahat More
graph = ox.graph_from_point(location_point, dist=750, dist_type="bbox", network_type='drive')
```

```python
nodes, streets = ox.graph_to_gdfs(graph)
streets.head()
```

#Buildings

```python
location_point = (22.5176397,88.3658511)    # Gariahat More
#graph = ox.graph_from_point(location_point, dist=750, dist_type="bbox", network_type='drive')
tags = {'building':True}
building = ox.geometries_from_point(location_point,tags, dist=750)
```

```python
cols = ['amenity','building', 'name']
building[cols].head()
```

```python
#buildings = building[building.geom_type == 'Polygon'][:1000]
buildings = building[building.geom_type == 'Polygon'][:100]
```

```python
#m = folium.Map([22.5176397,88.3658511], zoom_start=18, tiles='CartoDb dark_matter')
m = folium.Map([22.5176397,88.3658511], zoom_start=16)
#folium.GeoJson(buildings[:1000]).add_to(m)
folium.GeoJson(buildings[:100]).add_to(m)
m
```
