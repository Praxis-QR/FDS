```
my_map = Basemap(projection='ortho', lat_0=22, lon_0=88,
              resolution='l', area_thresh=1000.0)
 
my_map.drawcoastlines()
my_map.drawcountries()   # India map is not as per Govt of India
my_map.drawmapboundary(fill_color='aqua') 
my_map.fillcontinents(color='coral')

my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))
 
plt.show()
```

##Specific States

in the following two cells, we have used two different shapefile collections. please note that depending on the names used in the files there are small changes in the parameters used.

```
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.patches import PathPatch
import numpy as np

fig     = plt.figure()
ax      = fig.add_subplot(111)

#map3 = Basemap(projection='merc',llcrnrlon=68.1, llcrnrlat= 6.5, urcrnrlon=97.4, urcrnrlat=38)
map3 = Basemap(projection='merc',llcrnrlon=82, llcrnrlat= 20, urcrnrlon=92, urcrnrlat=28, resolution='h')


map3.drawmapboundary(fill_color='aqua')
#map3.fillcontinents(color='#ddaa66',lake_color='aqua')
map3.fillcontinents(color='sienna',lake_color='aqua')
map3.drawcoastlines()


map3.readshapefile('india2/INDIA', 'INDIA', drawbounds = False)
#map.readshapefile('../sample_files/comarques', 'comarques', drawbounds = False)

patches   = []

for info, shape in zip(map3.INDIA_info, map3.INDIA):
    if info['ST_NAME'] == 'West Bengal':
        patches.append( Polygon(np.array(shape), True) )
        
#ax.add_collection(PatchCollection(patches, facecolor= 'm', edgecolor='k', linewidths=1., zorder=2))
ax.add_collection(PatchCollection(patches, facecolor= 'lightgreen', edgecolor='k', linewidths=1., zorder=2))

plt.show()
```

```
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.patches import PathPatch
import numpy as np

fig     = plt.figure()
ax1      = fig.add_subplot(111)

#map3 = Basemap(projection='merc',llcrnrlon=68.1, llcrnrlat= 6.5, urcrnrlon=97.4, urcrnrlat=38)
map3 = Basemap(projection='merc',llcrnrlon=82, llcrnrlat= 20, urcrnrlon=92, urcrnrlat=28, resolution='h')


map3.drawmapboundary(fill_color='aqua')
#map3.fillcontinents(color='#ddaa66',lake_color='aqua')
map3.fillcontinents(color='olive',lake_color='azure')
map3.drawcoastlines()


map3.readshapefile('india2/INDIA', 'INDIA', drawbounds = False)
#map.readshapefile('../sample_files/comarques', 'comarques', drawbounds = False)

patches1   = []
patches2   = []

for info, shape in zip(map3.INDIA_info, map3.INDIA):
    if info['ST_NAME'] == 'West Bengal':
        patches1.append( Polygon(np.array(shape), True) )
    if info['ST_NAME'] == 'Jharkhand':
        patches2.append( Polygon(np.array(shape), True) )
        
#ax1.add_collection(PatchCollection(patches1, facecolor= 'blue', edgecolor='k', linewidths=1., zorder=2))
ax1.add_collection(PatchCollection(patches1, facecolor= 'lightblue', edgecolor='k', linewidths=1., zorder=2))
#ax1.add_collection(PatchCollection(patches2, facecolor= 'red', edgecolor='k', linewidths=1., zorder=2))
ax1.add_collection(PatchCollection(patches2, facecolor= 'yellow', edgecolor='k', linewidths=1., zorder=2))

plt.show()
```

# Plotting Cities

```
# Calcutta and Ranchi being plotted
lats = [22.5726,23.3441]
lons = [88.3639,85.3096]
```

```
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.patches import PathPatch
import numpy as np

fig     = plt.figure()
ax1      = fig.add_subplot(111)

#map3 = Basemap(projection='merc',llcrnrlon=68.1, llcrnrlat= 6.5, urcrnrlon=97.4, urcrnrlat=38)
map3 = Basemap(projection='merc',llcrnrlon=82, llcrnrlat= 20, urcrnrlon=92, urcrnrlat=28, resolution='h')


map3.drawmapboundary(fill_color='aqua')
#map3.fillcontinents(color='#ddaa66',lake_color='aqua')
map3.fillcontinents(color='olive',lake_color='azure')
map3.drawcoastlines()

x,y = map3(lons, lats)
map3.plot(x, y, 'ko', markersize=6)
#map3.plot(x, y, color='red', marker='o', markersize=8)
#plot(x, y, color='green', linestyle='dashed', marker='o',     markerfacecolor='blue', markersize=12).

labels = ['Kolkata', 'Ranchi']
for label, xpt, ypt in zip(labels, x, y):
    plt.text(xpt, ypt, label,fontsize=10)


map3.readshapefile('india2/INDIA', 'INDIA', drawbounds = False)
#map.readshapefile('../sample_files/comarques', 'comarques', drawbounds = False)

patches1   = []
patches2   = []

for info, shape in zip(map3.INDIA_info, map3.INDIA):
    if info['ST_NAME'] == 'West Bengal':
        patches1.append( Polygon(np.array(shape), True) )
    if info['ST_NAME'] == 'Jharkhand':
        patches2.append( Polygon(np.array(shape), True) )
        
#ax1.add_collection(PatchCollection(patches1, facecolor= 'blue', edgecolor='k', linewidths=1., zorder=2))
ax1.add_collection(PatchCollection(patches1, facecolor= 'lightblue', edgecolor='k', linewidths=1., zorder=2))
#ax1.add_collection(PatchCollection(patches2, facecolor= 'red', edgecolor='k', linewidths=1., zorder=2))
ax1.add_collection(PatchCollection(patches2, facecolor= 'yellow', edgecolor='k', linewidths=1., zorder=2))

plt.show()
```
